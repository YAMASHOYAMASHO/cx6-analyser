```python
from tkinter import filedialog
from matplotlib import pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [5, 4]
plt.rcParams['svg.fonttype'] = 'none'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.width'] = 1.5  # X軸の目盛り線の幅
plt.rcParams['ytick.major.width'] = 1.5  # Y軸の目盛り線の幅
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.bottom'] = False
plt.rcParams['axes.spines.left'] = False
plt.rcParams['font.size'] = 18
plt.rcParams['font.family'] = ['Arial', 'Meiryo']  # 優先順位で指定
```

```python
path = filedialog.askopenfilename()
```

```python
with open(path, 'r') as f:
    lines = f.readlines()
    print(lines)

MNAT = []
date = []
time = []
interval = []
temp_1 = []
temp_2 = []
temp_3 = []
temp_4 = []
unit = []

for line in lines[1:-1]:
    MNAT.append(line[0:2].strip())
    date.append(line[4:14].strip())
    time.append(line[15:23].strip())
    interval.append(line[24:29].strip())
    temp_1.append(float(line[30:36].strip()))
    # temp_2.append(float(line[36:44].strip()))
    temp_3.append(float(line[44:52].strip()))
    temp_4.append(float(line[52:60].strip()))
    unit.append(line[60:].strip())
```

```python
time = np.arange(len(temp_1)) * 5 / 60 / 60 # hours
```

```python
%matplotlib tk
```

```python
plt.plot(time, temp_1, label="空間")
# plt.plot(time, temp_2, label="T2")
plt.plot(time, temp_3, label="EP上")
plt.plot(time, temp_4, label="PEEK下")
plt.legend()
plt.xlabel("Time (h)")
plt.ylabel("Temperature (°C)")
plt.ylim(20, 200)
plt.yticks(np.arange(20, 201, 40))
plt.xlim(0, 9)
plt.savefig(filedialog.asksaveasfilename(defaultextension=".svg"))
```
